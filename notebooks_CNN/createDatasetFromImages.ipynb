```python
import tensorflow as tf
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image
import os
```

```python
dir_path = r'C:\Users\Kleinau\Desktop\BMW_neu'#Hier muss der Pfad zu den ordnern mit Namen S_6, S_1.323, etc. angegeben werden 
output_dir = os.path.join(r"..\data\paper_data\final_dataset")    #Dieser Ausgabeordner sollte vor Ausführung des Notebooks leer und vorhanden sein!
counter = 1  # Bild ID
image_exts = ['jpeg', 'jpg', 'bmp', 'png']
```

```python
def is_valid_image(image_path):
    try:
        img = Image.open(image_path)
        img.verify()  # Überprüft, ob das Bild beschädigt ist
        return True
    except Exception:
        return False
```

```python
for image in os.listdir(dir_path):
    image_path = os.path.join(dir_path, image)  # Pfad zum Bild

    # Überprüfen, ob es sich um ein gültiges Bild handelt
    if is_valid_image(image_path):
        print(f"Processing: {image_path}")

        # Extrahiere die Teile des Bildnamens
        parts = image.split('_')  # Annahme: Der Dateiname hat das Format 'AC_33_B2.jpg'

        # Extrahiere den SDAS-Wert und berechne den angepassten Wert
        try:
            sdas_value_str = parts[1]  # Das gibt '33' in 'AC_33_B2.jpg'
            SDAS = float(sdas_value_str)  # Den String '33' in eine Zahl umwandeln
            s_value = round(SDAS * (1 / 0.9904415111515698), 2)  # Die Formel anwenden
            print(f"Image: {image}, SDAS: {SDAS}, Adjusted Value: {s_value}")
        except ValueError:
            print(f"Invalid number in filename {image}")
            continue  # Weiter zur nächsten Bilddatei, wenn der Wert nicht umgewandelt werden kann

        # Bild in Graustufen einlesen
        img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        # Bilddimensionen
        height, width = img.shape
        min_size = 200  # Mindestgröße der Kacheln
        rows = height // min_size
        cols = width // min_size

        # Bild in Kacheln zerschneiden und speichern
        for i in range(rows):
            for j in range(cols):
                y_start = i * min_size
                x_start = j * min_size
                y_end = min(y_start + min_size, height)
                x_end = min(x_start + min_size, width)

                tile = img[y_start:y_end, x_start:x_end]
                tile_path = os.path.join(output_dir, f'S_{s_value}_ID_{counter}.png')  # Speichern der Kachel
                cv2.imwrite(tile_path, tile)

                counter += 1

#                 # 90°-Rotation und Speicherung
#                 rotated_tile = cv2.rotate(tile, cv2.ROTATE_90_CLOCKWISE)
#                 rotated_tile_path = os.path.join(output_dir, f'S_{s_value}_ID_{counter}.png')
#                 cv2.imwrite(rotated_tile_path, rotated_tile)

#                 counter += 1

#                 # 180°-Rotation und Speicherung
#                 rotated_tile = cv2.rotate(rotated_tile, cv2.ROTATE_90_CLOCKWISE)
#                 rotated_tile_path = os.path.join(output_dir, f'S_{s_value}_ID_{counter}.png')
#                 cv2.imwrite(rotated_tile_path, rotated_tile)

#                 counter += 1

#                 # 270°-Rotation und Speicherung
#                 rotated_tile = cv2.rotate(rotated_tile, cv2.ROTATE_90_CLOCKWISE)
#                 rotated_tile_path = os.path.join(output_dir, f'S_{s_value}_ID_{counter}.png')
#                 cv2.imwrite(rotated_tile_path, rotated_tile)

#                 counter += 1
    else:
        print(f"ERROR: Removing invalid image {image_path}")
```
