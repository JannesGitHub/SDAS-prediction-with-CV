Dieses Notebook ermöglicht die Erstellung eines Datensatzes im Stil der CNN-Studie. Zunächst müssen alle Bilder mit dem entsprechenden S gelabelt werden und in ihre Ordner sortiert werden. Gerade sind die Materialien noch getrennt, aber eigentlich würde man später nur nach S klassifizieren.
Die Trainingsbilder müssen also in Ordnern mit den Namen "S_{S Wert hier einfügen vorliegen}" und alle im gleichen Überordner liegen. Der erstellte Datensatz besteht aus grauskalierten 200x200-Bildern die mit dem namen "S_{S Wert hier einfügen vorliegen}_ ID _{ID hier einfügen}" danach vorliegen und somit direkt für die trainingspipeline verwendet werden können.

```python
import tensorflow as tf
import numpy as np
import cv2
from matplotlib import pyplot as plt
from PIL import Image
import os
```

```python
dir_path = r'C:\Users\Kleinau\Desktop\S_Dendrite_CNN_Daten\raw_data'#Hier muss der Pfad zu den ordnern mit Namen S_6, S_1.323, etc. angegeben werden 
output_dir = os.path.join(r"..\data\paper_data\final_dataset")               #Dieser Ausgabeordner sollte vor Ausführung des Notebooks leer und vorhanden sein!
counter = 1  # Bild ID
classes = os.listdir(dir_path) 
image_exts = ['jpeg', 'jpg', 'bmp', 'png']
```

```python
os.listdir(dir_path)
```

```python
def is_valid_image(image_path):
    try:
        img = Image.open(image_path)
        img.verify()  # Überprüft, ob das Bild beschädigt ist
        return True
    except Exception:
        return False
```

```python
for cls in classes:
    cls_path = os.path.join(dir_path, cls)
    
    images = os.listdir(cls_path)
    
    s_value = round(float(cls), 3)
    
    for image in images:
        image_path = os.path.join(cls_path, image)

        if is_valid_image(image_path):
            print(f"Processing: {image_path}")

            # Bild in Graustufen einlesen
            img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

            # Bilddimensionen
            height, width = img.shape
            min_size = 200  # Mindestgröße der Kacheln
            rows = height // min_size
            cols = width // min_size

            # Bild in Kacheln zerschneiden und speichern
            for i in range(rows):
                for j in range(cols):
                    y_start = i * min_size
                    x_start = j * min_size
                    y_end = min(y_start + min_size, height)
                    x_end = min(x_start + min_size, width)

                    tile = img[y_start:y_end, x_start:x_end]
                    tile_path = os.path.join(output_dir, f'S_{s_value}_ID_{counter}.png') #eigentlich brauchts nur s_value und id
                    cv2.imwrite(tile_path, tile)

                    counter += 1

                    # 90°-Rotation und Speicherung
                    rotated_tile = cv2.rotate(tile, cv2.ROTATE_90_CLOCKWISE)
                    rotated_tile_path = os.path.join(output_dir, f'S_{s_value}_ID_{counter}.png')
                    cv2.imwrite(rotated_tile_path, rotated_tile)

                    counter += 1
                    
                    # 180°-Rotation und Speicherung
                    rotated_tile = cv2.rotate(rotated_tile, cv2.ROTATE_90_CLOCKWISE)
                    rotated_tile_path = os.path.join(output_dir, f'S_{s_value}_ID_{counter}.png')
                    cv2.imwrite(rotated_tile_path, rotated_tile)

                    counter += 1
                    
                    # 270°-Rotation und Speicherung
                    rotated_tile = cv2.rotate(rotated_tile, cv2.ROTATE_90_CLOCKWISE)
                    rotated_tile_path = os.path.join(output_dir, f'S_{s_value}_ID_{counter}.png')
                    cv2.imwrite(rotated_tile_path, rotated_tile)
                    
                    counter += 1
        else:
            print(f"ERROR: Removing invalid image {image_path}")
```
